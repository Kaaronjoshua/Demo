```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("50_Startups.csv")
```

```python
df
```

```python
df.rename(columns = {'Marketing Spend':'Marketing_Spend'}, inplace = True)
```

```python
X=df.iloc[:,1:3]
X
```

```python
Y=df.iloc[:,-1]
Y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg= LinearRegression()
reg.fit(X,Y)
```

```python
X_surf,Y_surf=np.meshgrid(np.linspace(X.Administration.min(),X.Administration.max(),100),np.linspace(X.Marketing_Spend.min(),X.Marketing_Spend.max(),100))
```

```python
onlyx=pd.DataFrame({'Administration':X_surf.ravel(),'Marketing_Spend':Y_surf.ravel()})
fittedY=reg.predict(onlyx)
```

```python
fittedY=np.array(fittedY)
```

```python
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(X['Marketing_Spend'],X['Administration'],df['Profit'],c='red',marker='+',alpha=0.5)
ax.plot_surface(X_surf,Y_surf,fittedY.reshape(X_surf.shape),color='None',alpha=0.3)
ax.set_xlabel('Administration')
ax.set_ylabel('Market_spend')
ax.set_zlabel('Price')
plt.show()
```
