```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df=pd.read_csv("Knn2.csv")
```

```python
df
```

```python
df.corr()
```

```python
dataplot = sb.heatmap(df.corr(), annot=True)
```

```python
x1=df.iloc[:,[-1,-8]].values
x1
```

```python
from sklearn.cluster import KMeans
wcss=[]
for k in range(1,11):
    kmeans= KMeans(n_clusters=k,init='k-means++',random_state=42)
    kmeans.fit(x1)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.xlabel('clusters')
plt.ylabel('wcss')
plt.title('Elbow Graph')
plt.show()
```

```python
kmeans= KMeans(n_clusters=3,init='k-means++',random_state=42)
y=kmeans.fit_predict(x1)
```

```python
plt.scatter(x1[y==0,0],x1[y==0,1],c='red',s=100,label='Cluster-1')
plt.scatter(x1[y==1,0],x1[y==1,1],c='blue',s=100,label='Cluster-2')
plt.scatter(x1[y==2,0],x1[y==2,1],c='green',s=100,label='Cluster-3')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='yellow',s=300,label='Center')
plt.legend()
plt.xlabel('price_range')
plt.ylabel('ram')
plt.show()
```
