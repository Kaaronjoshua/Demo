```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def bad_line(x):
    print(x)
    return None
df = pd.read_csv('groceries.csv', encoding='unicode_escape', on_bad_lines=bad_line, engine='python',header=None)
print(df)
```

```python
df.shape
```

```python
transactions=[]
for i in range(0,6106):
    transactions.append([str(df.values[i,j])for j in range(0,4)])
```

```python
transactions[:2]
```

```python
from itertools import permutations
flattened = [item for transaction in transactions for item in transaction]
items = list(set(flattened))
```

```python
rules = list(permutations(items, 2))
print(rules[:10])
```

```python
print(len(rules))
```

```python
from apyori import apriori
rules=apriori(transactions=transactions,min_support=0.0003,min_confidence=0.2,min_lift=3,min_length=2,max_length=2)
```

```python
results=list(rules)
```

```python
results
```

```python
def inspect(results):
    lhs=[tuple(result[2][0][0])[0] for result in results]
    rhs=[tuple(result[2][0][1])[0] for result in results]
    supports=[result[1] for result in results]
    confidence=[result[2][0][2] for result in results]
    lifts=[result[2][0][3] for result in results]
    return list(zip(lhs,rhs,supports,confidence,lifts))
df=pd.DataFrame(inspect(results),columns=['Product-1','Product-2','Support','Confidence','Lift'])
```

```python
df
```

```python
df.nlargest(n=5,columns='Lift')
```

```python
rules_df = df.sort_values('Support', ascending=False).head(200)

# List of all items
items = set(df['Product-1']) | set(rules_df['Product-2'])

# Creates a mapping of items to numbers
imap = {item : i for i, item in enumerate(items)}

# Maps the items to numbers and adds the numeric 'FromN' and 'ToN' columns
df['Product-1N'] = df['Product-1'].map(imap)
df['Product-2N'] = df['Product-2'].map(imap)

# Displays the top 20 association rules, sorted by Support
df.head(20)
```
